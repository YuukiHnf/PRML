```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy import linalg as LA
import matplotlib.cm as cm
plt.style.use("ggplot")
```

```
x_ = np.linspace(-3,10,100)
def f(x):
  return 0.5 * stats.norm.pdf(x, 0, 1) + 0.5 * stats.norm.pdf(x, 5, 2)

plt.plot(x_, f(x_))
plt.show()
```

```
def max_min(x, y):
  delta = 0.5
  i = 0
  while f(x + delta*i) > y:
    i += 1
  x_max = x + delta * i
  j = 0
  while f(x - delta*i) > y:
    i += 1
  x_min = x - delta * j
  return x_min, x_max

x = 1
X = []
for i in range(5000):
  y = stats.uniform(0,f(x)).rvs(1)[0]
  x_min, x_max = max_min(x,y)
  while True:
    new_x = stats.uniform(x_min, x_max-x_min).rvs(1)[0]
    if f(new_x) > y:
      x = new_x
      X.append(x)
      break

plt.hist(X, bins=50, density=True)
plt.plot(x_, f(x_))
plt.show()
```
